```python
import numpy as np
from cv2 import imread
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))
```

```python
Img=imread('flowers.jpg') #BGR
cv2_imshow(Img)
howis(Img)
```

```python
I=Img[:,:,[2,1,0]] #BGR to RGB
plt.imshow(I)
```

```python
#CANALES
R=I [:,:,0]
G=I [:,:,1]
B=I [:,:,2]
RGB=np.concatenate((R,G,B), axis=0) #1 es horizontal y 0 vertical
plt.imshow(RGB, cmap='gray')
```

```python
#0-255
Sr= R>180
Sg= G<50
Sb= B<50
Srgb=np.concatenate((Sr,Sg,Sb), axis=1)
plt.imshow(Srgb, cmap='gray')
```

```python
Suma=np.logical_and(Sr,Sg)
cv2_imshow(Suma*255)
```

```python
Suma2=np.logical_and(Suma,Sb)
cv2_imshow(Suma*255)
```

Para descomponer imagen y seguir seccionando

```python
(N,M)=Suma2.shape
Q=Suma2

for i in range(N):
    for j in range(M):
        s=np.sum(Suma2[i,:])
        if s<40:
            Q[i,:]=0

cv2_imshow(Q*255)
```
